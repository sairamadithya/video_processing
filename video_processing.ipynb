# video processing

```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture(0) 
fourcc = cv2.VideoWriter_fourcc('X','V','I','D') 
out = cv2.VideoWriter('op1.avi', fourcc, 8.0, (640,480))
print("VIDEO PROCESSING TUTORIAL")
print("1.original 2.grayscale 3.RGB 4.blue 5.green 6.red 7.negative 8.grayscale_negative 9.canny edge 10.dilated edge")
print("11.eroded edge 12.autumn 13.bone 14.jet 15.winter 16.rainbow 17.ocean 18.summer 19.spring 20.cool ")
print("21.hsv 22.pink 23.hot 24.parula 25.magma 26.inferno 27.plasma 28.viridis 29.cividis 30.twilight")
print("31.twilight_shifted 32.turbo  33.deepgreen 34.flipped image 35.lab image")
print("Enter the number of choice from 1 to 35")
choice=int(input())
while True:
    success, frame = cap.read()#frame variable will capture the Video & success variable will tell us whether it was captured successfully or not
    if choice==1:
        frames= frame
    elif choice==2:
        frames=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    elif choice==3:
        frames=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    elif choice==4:
        b, g, r = cv2.split(frame)
        zeros_ch = np.zeros(frame.shape[0:2], dtype="uint8")
        frames = cv2.merge([b, zeros_ch, zeros_ch])
    elif choice==5:
        b, g, r = cv2.split(frame)
        zeros_ch = np.zeros(frame.shape[0:2], dtype="uint8")
        frames = cv2.merge([ zeros_ch,g, zeros_ch])    
    elif choice==6:
        b, g, r = cv2.split(frame)
        zeros_ch = np.zeros(frame.shape[0:2], dtype="uint8")
        frames = cv2.merge([ zeros_ch, zeros_ch,r]) 
    elif choice==7:
        frames=cv2.bitwise_not(frame)
    elif choice==8:
        frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        frames=cv2.bitwise_not(frame)
    elif choice==9:
        frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        frames=cv2.Canny(frame,140,190)
    elif choice==10:
        frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        canny=cv2.Canny(frame,125,175)
        kernel=np.ones((3,3),np.uint8)
        frames=cv2.dilate(canny,kernel,iterations=1)
    elif choice==11:
        frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        canny=cv2.Canny(frame,125,175)
        kernel=np.ones((1,1),np.uint8)
        frames=cv2.erode(canny,kernel,iterations=1)   
    elif choice==12:
        frames=cv2.applyColorMap(frame,0)
    elif choice==13:
        frames=cv2.applyColorMap(frame,1)
    elif choice==14:
        frames=cv2.applyColorMap(frame,2)
    elif choice==15:
        frames=cv2.applyColorMap(frame,3)
    elif choice==16:
        frames=cv2.applyColorMap(frame,4)
    elif choice==17:
        frames=cv2.applyColorMap(frame,5)
    elif choice==18:
        frames=cv2.applyColorMap(frame,6)
    elif choice==19:
        frames=cv2.applyColorMap(frame,7)
    elif choice==20:
        frames=cv2.applyColorMap(frame,8)
    elif choice==21:
        frames=cv2.applyColorMap(frame,9)
    elif choice==22:
        frames=cv2.applyColorMap(frame,10)
    elif choice==23:
        frames=cv2.applyColorMap(frame,11)
    elif choice==24:
        frames=cv2.applyColorMap(frame,12)
    elif choice==25:
        frames=cv2.applyColorMap(frame,13)
    elif choice==26:
        frames=cv2.applyColorMap(frame,14)
    elif choice==27:
        frames=cv2.applyColorMap(frame,15)
    elif choice==28:
        frames=cv2.applyColorMap(frame,16)
    elif choice==29:
        frames=cv2.applyColorMap(frame,17)
    elif choice==30:
        frames=cv2.applyColorMap(frame,18)
    elif choice==31:
        frames=cv2.applyColorMap(frame,19)
    elif choice==32:
        frames=cv2.applyColorMap(frame,20)
    elif choice==33:
        frames=cv2.applyColorMap(frame,21)
    elif choice==34:
        frames=cv2.flip(frame,0)
    elif choice==35:
        frames=cv2.cvtColor(frame,cv2.COLOR_BGR2LAB)
                                                
                                                    
        
    if success == True:
        
        out.write(frame)
        
        cv2.imshow("Video", frames)
    
        if cv2.waitKey(1) == ord('q'): #This adds a Delay and looks for the key press inorder to break the loop
            break
            
    else:
        break
        
print("Press q in the keyboard to stop the video")        
cap.release() #Release the resources after recording
out.release()
cv2.destroyAllWindows()
```
